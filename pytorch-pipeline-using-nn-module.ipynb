```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
    def __init__(self, num_feature):
        super().__init__()
        self.linear = nn.Linear(num_feature, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)

        return out
        
```

```python
features  = torch.rand(10, 5)

model = Model(features.shape[1])
model(features)
```

```python
class Model_NN(nn.Module):
    def __init__(self, num_feature):
        super().__init__()
        self.linear1 = nn.Linear(num_feature, 3)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(3 ,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear1(features)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)

        return out
        
```

```python
features  = torch.rand(10, 5)

model = Model_NN(features.shape[1])
model(features)
```

# Sequential container

```python
class Model_NN_Seq(nn.Module):
    def __init__(self, num_feature):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_feature, 3),
            nn.ReLU(),
            nn.Linear(3, 1),
            nn.Sigmoid()
        )

    def forward(self, features):
        out = self.network(features)

        return out
```

```python
features  = torch.rand(10, 5)

model_ = Model_NN_Seq(features.shape[1])
model(features)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor = X_train_tensor.float()  # Convert to float32
y_train_tensor = y_train_tensor.float()  # If labels are floats (e.g., for regression)
y_test_tensor = y_test_tensor.float()
X_test_tensor = X_test_tensor.float()
```

```python
class RWModel(nn.Module):
    def __init__(self, num_feature):
        super().__init__()
        self.linear = nn.Linear(num_feature, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)

        return out

    # def loss_function(self, y_pred, y):
    #     # Clamp predictions to avoid log(0)
    #     epsilon = 1e-7
    #     y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    
    #     # Calculate loss
    #     loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    #     return loss
```

```python
learning_rate = 0.1
epochs = 25
```

```python
loss_function = nn.BCELoss()
```

```python
model = RWModel(X_train_tensor.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model(X_train_tensor)

  # loss calculate
  loss = loss_function(y_pred, y_train_tensor.view(-1 , 1))


  optimizer.zero_grad()
  # backward pass
  loss.backward()

  # # parameters update
  # with torch.no_grad():
  #   model.linear.weight -= learning_rate * model.linear.weight.grad
  #   model.linear.bias -= learning_rate * model.linear.bias.grad

  optimizer.step()
  
    
  # zero gradients
  # model.linear.weight.grad.zero_()
  # model.linear.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > .9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
