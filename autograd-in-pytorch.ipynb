# Neural Network

```python
import torch 

x = torch.tensor(6.7) # input feature
y = torch.tensor(0.0) # True label


w = torch.tensor(1.0) # weight
b = torch.tensor(0.0) # bias
```

```python
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8
    prediction = torch.clamp(prediction, epsilon , 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 -prediction))
```

```python
# Forward pass
z = w * x + b # weight sum (linear part)
y_pred = torch.sigmoid(z) # predication probability

# compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
```

```python
loss
```

### Backpropagation with Manual calculate derivatives

```python
# derivatives
# 1. dL/d(y_pred): loss with respect to prediction (y_pred)
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))

# 2. dy_pred/dz: predication (y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

# 3. dz/dw and dz/db: z with respect to w and b
dz_dw = x  # dz/dw = x
dz_db = 1 # dz/db = 1 (bise contributes directly to z)

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f'Manual Gradient of loss weight (dw): {dL_dw}')
print(f'Manual Gradient of loss bias (dw): {dL_db}')
```

### Backpropagation with grad calculate derivatives

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
print(w, b)
```

```python
# farword pass
z = w * x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad, b.grad)
```
